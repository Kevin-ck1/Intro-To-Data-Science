### Next Steps: Data Cleaning Steps

**Error in values**

- Extract the first integer value (lowest delay time) in the column "How_Long_Delayed"


**Missing Values**

- Impute the missing values in the field "How_Long_Delayed" with the mean value.


**Error in Datatypes**

- Convert "How_Long_Delayed" to int datatype.



**Error in field names**

- Rename the column "Boro" to "Borough".

```
# Lets first start by creating a copy of our dataframe 
# df_clean = df.copy(). We will use this copy as our cleaning copy.
# ---
#
bb_clean = bb.copy()
bb_clean.head()
```

```
# Then extracting the lowest delay time in the column how_long_delayed from the string
# 
bb_clean['how_long_delayed'] = bb_clean['how_long_delayed'].str.extract('(\d+)')
bb_clean['how_long_delayed'].head()
```

We impute the null values in 'how_long_delayed' column with mean of the column. This will take a couple of steps...

```
# We first convert our how_long_delayed to float type to allow for imputation
# ---
# 

bb_clean["how_long_delayed"] = bb_clean["how_long_delayed"].apply(np.float64)

bb_clean["how_long_delayed"].dtypes
```

```
# Then later perform our mean imputation 
# ---
# 

#bb_clean["how_long_delayed"].mean(skipna = True)--- TO find the mean 

# bb_clean["how_long_delayed"] = bb_clean["how_long_delayed"].fillna(np.nan,(bb_clean["how_long_delayed"].mean(skipna = True)))

# .fillna con also be used instead of .replace


bb_clean["how_long_delayed"] = bb_clean["how_long_delayed"].replace(np.nan,(bb_clean["how_long_delayed"].mean(skipna = True)))
bb_clean["how_long_delayed"]
```

```
# Then convert back our how_long_delayed column to integer datatype
# ---
# 

bb_clean["how_long_delayed"] = bb_clean["how_long_delayed"].apply(np.int64)

bb_clean["how_long_delayed"]

#bb_clean["how_long_delayed"].dtypes
```

```
# Then check for nulls in the column
# ---
# 
bb_clean["how_long_delayed"].isnull().any()
```

```
# Rename Boro column to Borough
# ---
# 

bb_clean.rename(columns={"boro":"borough"}, inplace = True)

bb_clean.head()
```

```
# Lastly we convert all values in our colums to lower case
# for ease of reading
# ---
# 
'''
bb_lowercase = bb_clean['school_year', 'busbreakdown_id', 'run_type', 'bus_no', 'route_number',
       'reason', 'schools_serviced', 'occurred_on', 'created_on', 'borough',
       'bus_company_name', 'how_long_delayed', 'number_of_students_on_the_bus',
       'has_contractor_notified_schools', 'has_contractor_notified_parents',
       'have_you_alerted_opt', 'informed_on', 'incident_number',
       'last_updated_on', 'breakdown_or_running_late', 'school_age_or_prek'].str.lower()
'''
#Only values in the borough are converted to lowercase

#bb_clean['borough'] = bb_clean["borough"].str.lower()

bb_lower = bb_clean.applymap(lambda x: x.lower() if type(x)== str else x)
bb_lower.head()
```

```
# Check the first 5 record the cleaned dataset
# ---
#
bb_clean = bb_lower
bb_clean.head()
```


## 5. Solution Implementation

Here we investigate the questions that would help craft our recommendations.

### 5.a) Questions

```
# 1. Which bus companies that had the highest breakdowns?
# ---
#
breakdowns = bb_clean.groupby(['bus_company_name']).count().reset_index()

# Sort to get bus company with highest breakdowns
# ---
# YOUR CODE GOES BELOW
#
breakdowns.sort_values(by='school_year', ascending=0)
```

```
# 2. What were the top 3 reasons for bus delays?
# ---
#
bus_delay = bb_clean.groupby(['reason']).count()[['how_long_delayed']].reset_index()

# Sort to get the most frequent reason
# ---
# YOUR CODE GOES BELOW
#

bus_delay.sort_values(by ='how_long_delayed', ascending = 0)

#bus_delay
```

```
# 3. How many students were in the buses when they broke down? 
# ---
#
students = bb_clean.groupby(['number_of_students_on_the_bus']).count()[['busbreakdown_id']].reset_index()

students.sample(20)
```

```
# 4. Which were most frequent reasons for bus breakdowns?
# ---
#
breakdown_reasons = bb_clean[bb_clean.breakdown_or_running_late == 'breakdown'].groupby(['reason']).count()

# Sort to get most frequent reasons
# ---
# YOUR CODE GOES BELOW
#
breakdown_reasons.sort_values(by="school_year", ascending=0)
```

```
# 5. What were the most frequent reasons for the bus running late? 
# ---
#

#reasons = bb_clean.groupby(['reason','how_dong_delayed']).count()


# Get the records with running late reasons and sort to get most frequent reasons
# ---
# YOUR CODE GOES BELOW
#

reasons = bb_clean[bb_clean.breakdown_or_running_late == 'running late'].groupby(['reason']).count()[['how_long_delayed']].reset_index()
reasons.sort_values(by="how_long_delayed", ascending=0)
```

```
# 6. What was the average delay time for each reason type?
# ---
# 

# Get the records with reasons and how long on average a delay took then sort 
# ---
# YOUR CODE GOES BELOW
#
avg_delay = bb_clean.groupby('reason').mean()[['how_long_delayed']].reset_index()

avg_delay.sort_values(by="how_long_delayed", ascending=0)
```
